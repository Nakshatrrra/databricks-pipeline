```python
from pyspark.sql.functions import *
```

```python
silver_df = spark.table("silver_orders")
```

```python
sales_by_category = (
    silver_df
    .groupBy("category")
    .agg(
        sum("total_amount")
        .alias("revenue")
    )
)
```

```python
sales_by_category.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_sales_by_category")
```

```python
customer_revenue = (
    silver_df
    .groupBy(
        "customer_id",
        "customer_name"
    )
    .agg(
        sum("total_amount")
        .alias("total_spend")
    )
)
```

```python
customer_revenue.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_customer_revenue")
```

```python
%sql
SELECT *
FROM gold_customer_revenue
ORDER BY total_spend DESC
```
